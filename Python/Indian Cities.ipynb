# Indian Cities

You are hired as a Data Analyst by Indian Government to investigate Top 500 Indian Cities statistical indicators.  
You have been supplied with a data set, and are asked to visually present the answers for following questions:

1. Which states have highest population count? 
2. Compare literacy across the states
3. Top 10 cities with high literacy
4. Explore sex ratio of different states
5. Percentage of total Graduate against Total Population of States
6. States with most cities in Top 500.


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 12,8
```

```python
df = pd.read_csv('/Users/Rudra/Documents/Work/Projects/SuperDataScience Workshops/Python/Indian Cities/IndianCities.csv')
```

```python
df.info()
```

```python
colour = ['darkred','darkgreen','silver','teal','pink','orange','salmon','plum','peru']
```

```python
state_population = df[['state_name','population_total']].groupby('state_name').sum().sort_values(['population_total'],ascending=False)
```

```python
state_population
```

---

## 1. Highest Population

```python
state_population.plot(kind='bar', color = colour)
plt.xlabel('State Name', size = 20)
plt.ylabel('Total Population', size = 20)
plt.show()
```

---

## 2. Literacy by State

```python
literacy_by_state = df[['effective_literacy_rate_total','state_name']].groupby('state_name').mean().sort_values(['effective_literacy_rate_total'],ascending=False)
literacy_by_state.plot(kind='bar', color = colour)
plt.xlabel('State Name', size = 20)
plt.ylabel('Avg Literacy Rate', size = 20)
plt.ylim(75,100)
plt.show()
```

---

## 3. Top 10 Literate Cities

```python
literacy_by_city = df[['effective_literacy_rate_total','city']].groupby('city').mean().sort_values(['effective_literacy_rate_total'],ascending=False)
top10 = literacy_by_city.head(10)
top10.plot(kind='bar', color = colour)
plt.xlabel('State Name', size = 20)
plt.ylabel('Avg Literacy Rate', size = 20)
plt.ylim(90,100)
plt.show()
```

---

## 4. Sex Ratio 

```python
sex_ratio_by_state = df[['sex_ratio','state_name']].groupby('state_name').mean().sort_values(['sex_ratio'],ascending=False)
sex_ratio_by_state.plot(kind='bar', color = colour)
plt.xlabel('State Name', size = 20)
plt.ylabel('Avg Sex Rate', size = 20)
plt.ylim(800,1100)
plt.show()
```

---

## 5. Percentage of Total Graduates against Total Population of States


```python
df['graduates_percentage'] = df.total_graduates*100/df.population_total
```

```python
df.head()
```

```python
graduates_by_state = df[['graduates_percentage','state_name']].groupby('state_name').mean().sort_values(['graduates_percentage'],ascending=False)
graduates_by_state.plot(kind='bar', color = colour)
plt.title('Percentage of Total Graduates against Total Population of States', size = 20)
plt.xlabel('State Name', size = 20)
plt.ylabel('Graduate Percentage by State', size = 20)
plt.ylim(7.5)
plt.show()
```

---

## 6. States with the most Cities

```python
number_of_cities = df[['city','state_name']].groupby('state_name').count().sort_values(['city'],ascending=False)

number_of_cities.plot(kind='bar',color=colour)
plt.xlabel('State')
plt.ylabel('No. of Cities')
plt.show()
```
